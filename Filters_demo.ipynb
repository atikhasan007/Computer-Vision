UnderStanding compositon of colored images


```python
def visualize_RGB_Channel(imgArray=None, figsize=(10,7)):
  # spliting the RGB components
  B,G,R = cv2.split(imgArray)

  # create zero matrix of shape of image
  zeros = np.zeros(B.shape, dtype=B.dtype) # can use any channel

  #init subplots
  fig, ax = plt.subplots(2,2,figsize=figsize)


  #plotting the actual image and RGB image
  [axi.set_axis_off() for axi in ax.ravel()]

  ax[0,0].set_title("Original Image ")
  ax[0,0].imshow(cv2.merge((R,G,B)))


  ax[0,1].set_title("Red Channel")
  ax[0,1].imshow(cv2.merge((R,zeros,zeros)))

  ax[1,0].set_title("Green Channel")
  ax[1,0].imshow(cv2.merge((zeros,G,zeros)))

  ax[1,1].set_title("Blue Channel")
  ax[1,1].imshow(cv2.merge((zeros,zeros,B)))
```

```python
def simple_conv(imgFilter=None, picture=None):

  # extract the shpae of the image
  p_row, p_col = picture.shape

  k = imgFilter.shape[0] # k = 3

  temp = list()
  stride = 1

  #resultant image size

  final_cols = (p_col - k)//stride + 1
  final_rows = (p_row - k)//stride + 1

  # take vertically down stride across row hy row
  for v_stride in range(final_rows):
    # take horizontal right stride across col by col
    for h_stride in range(final_cols):
      target_area_of_pic = picture[v_stride: v_stride + k, h_stride: h_stride + k]
      z = sum(sum(imgFilter * target_area_of_pic))
      temp.append(z)

  return np.array(temp).reshape(final_rows, final_cols)
```
